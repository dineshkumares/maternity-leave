```python
from lxml import html
import requests
```

```python
page = requests.get('https://fairygodboss.com/maternity-leave-resource-center?page=79')
tree = html.fromstring(page.content)
```

```python
print(tree)
```

```python
print(html.tostring(tree))
```

```python
rows = tree.xpath('//tr')
```

```python
# Remove white space
companies = rows[0].xpath('//td/text()')

split_companies = []

for x in companies:
    split_companies.append(x.split())
    
print(split_companies)
```

```python
# Each row is 8 cells long, so split by 8
separated_companies = []
one_company = []

counter = 0
for x in split_companies:
    one_company.append(x)
    counter+=1
    if counter == 8:
        separated_companies.append(one_company)
        one_company = []
        counter = 0

for x in separated_companies:
    print(x)
```

```python
def objectCompany(n): 
    return {'Company': ' '.join(n[0]), 'Industry': ' '.join(n[1]), 'MaternityLeave': n[2][0]}

# Organize elements into a dictionary
organized_companies = map(objectCompany, separated_companies)

print(list(organized_companies))
```
